## 4.1 Components of a quantum network

A quantum network is a communication system designed to send quantum information (qubits) between physically separated locations. Unlike classical networks that transmit bits (0s and 1s), quantum networks leverage principles like entanglement and superposition. They consist of two fundamental components: a quantum channel and a classical channel, which work together to enable quantum communication protocols.

### Quantum Channel
The quantum channel is the physical medium used to transmit quantum states, or qubits, from one node to another.

* Function: Its sole purpose is to move fragile quantum information. This could be the state of a single photon, an atom, or another quantum system.
* Physical Medium: Typically, this is an optical fiber or a free-space laser link through which photons are sent.
* Challenges: Quantum channels are highly susceptible to environmental interference, known as "noise," which can corrupt the quantum state (a process called decoherence). Unlike classical signals, a quantum state cannot be copied or amplified without destroying it (due to the no-cloning theorem). This makes transmitting qubits over long distances a major challenge.

### Classical Channel
The classical channel is a conventional communication link, identical to what we use for the internet today.

* Function: It is used to send classical bits of information between the nodes in the quantum network. This information is crucial for coordinating and completing quantum protocols.
* Use Cases: For example, after two nodes receive an entangled pair of qubits, they might need to perform measurements. The results of these measurements (which are classical data) are sent over the classical channel to allow the other party to interpret their own qubit or complete a protocol. It is also used for error correction and synchronization.
* Analogy: Think of the quantum channel as a special, secure courier carrying a fragile, locked box (the qubit). The classical channel is the telephone line used by the sender to call the receiver and give them the combination to the lock after the box has arrived. Both are needed to get the message across.

## 4.2 Steps of the Quantum Teleportation protocol.

Quantum teleportation is a protocol, not a form of transportation. It doesn't move matter; it moves quantum information. It allows for the perfect transfer of an unknown quantum state from a sender (traditionally named Alice) to a receiver (Bob) with the help of a shared entangled pair and a classical communication channel.

### The Core Idea
The state of Alice's original qubit is destroyed at her location and "reborn" at Bob's location. The information appears to have "teleported" from one place to another.

### Protocol Steps
Let's say Alice wants to teleport the state of a qubit, let's call it Qubit M (for Message), to Bob.

1. Step 1: Entanglement Distribution:
    * Alice and Bob must first share a pair of entangled qubits. Let's call Alice's half Qubit A and Bob's half Qubit B. This entangled pair is often created at a central source and sent to both of them over quantum channels.
2. Step 2: Alice Prepares for Measurement:
    * Alice now possesses two qubits: Qubit M (the message) and Qubit A (her half of the entangled pair). Bob is far away, holding only Qubit B.
3. Step 3: Bell State Measurement (BSM):
    * Alice performs a special joint measurement on her two qubits (M and A). This is called a Bell State Measurement. This measurement "entangles" her message qubit with her half of the shared pair.
    * This measurement has four possible outcomes, which can be represented by two classical bits of information (00, 01, 10, or 11).
    * The moment Alice performs her measurement, the state of Bob's qubit (B) instantly changes, but in a way that is useless without more information. The original state of Qubit M is destroyed in this process.
4. Step 4: Classical Communication:
    * Alice sends the two classical bits—the result of her measurement—to Bob over the classical channel.
5. Step 5: Bob's Reconstruction:
    * Bob receives the two classical bits from Alice. Depending on which of the four results he receives, he knows exactly what transformation (or "correction") he needs to apply to his qubit (B). He applies a specific quantum gate based on the message. For example, if he receives "01", he might apply an X-gate.

**Result**: After Bob applies the correct transformation, his Qubit B is now in the identical state that Alice's Qubit M was in at the beginning. The quantum information has been successfully teleported.
