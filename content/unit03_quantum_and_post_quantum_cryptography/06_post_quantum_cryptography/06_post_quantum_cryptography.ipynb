## 6.1 Threat that quantum computers pose to classical cryptography

The vast majority of today's secure communication—from online banking and e-commerce to secure messaging—is protected by public-key cryptography. The most common systems, like RSA and Elliptic Curve Cryptography (ECC), base their security on mathematical problems that are practically impossible for classical computers to solve.
* RSA's Hard Problem: Factoring a very large number into its two prime components.
* ECC's Hard Problem: Solving the elliptic curve discrete logarithm problem.

### The Quantum Threat: Shor's Algorithm

In 1994, mathematician Peter Shor developed a quantum algorithm that can solve both of these "hard" problems with astonishing efficiency. A sufficiently powerful quantum computer running Shor's algorithm could break current encryption standards in hours or days, rendering much of our current digital security infrastructure obsolete. This future risk is often called the "Quantum Threat."

## 6.2 Difference between QKD and PQC.

Both Quantum Key Distribution (QKD) and Post-Quantum Cryptography (PQC) are solutions to the quantum threat, but they are fundamentally different approaches.

| Feature              | Quantum Key Distribution (QKD)                                               | Post-Quantum Cryptography (PQC)                                             |
| -------------------- | ---------------------------------------------------------------------------- | --------------------------------------------------------------------------- |
| Underlying Principle | Physics-based. Security is guaranteed by the laws of quantum mechanics.      | Math-based. Relies on new mathematical problems hard for quantum computers. |
| Main Goal            | Securely **distribute** a secret key.                                        | **Replace** current public-key algorithms (like RSA and ECC).               |
| How it Works         | Transmits qubits; eavesdropping is detected by measuring disturbances.       | Runs on classical computers and networks; a software upgrade.               |
| Hardware Required    | Specialized quantum hardware (quantum channels, photon detectors, etc.).     | No new hardware needed. Can run on existing classical infrastructure.       |
| Analogy              | A high-tech courier system where any attempt to peek in the box is detected. | Designing a new type of lock that is too complex for any known key-picker.  |

## 6.3 Introducing PQC algorithms

Post-Quantum Cryptography algorithms are classical algorithms designed to run on the computers we use today, but they are built on mathematical problems believed to be secure against attacks from both classical and quantum computers.

One of the most promising families of PQC algorithms is based on lattices.

### What is a Lattice?
Imagine a perfectly repeating grid of points, like the atoms in a crystal or a perfectly tiled floor, extending infinitely in many dimensions. This is a lattice.

### The Hard Problem: Closest Vector Problem
Lattice-based cryptography derives its security from problems that are hard to solve within these structures. A key example is the Closest Vector Problem (CVP).

Imagine you are given the coordinates of a single point somewhere near a massive, 1000-dimensional lattice. The challenge is to find the single lattice point that is closest to your given point. While this sounds easy in two dimensions, it becomes exponentially difficult in high dimensions, even for a quantum computer.

### CRYSTALS-Kyber
Kyber is a specific lattice-based algorithm that has been selected by the U.S. National Institute of Standards and Technology (NIST) as a new standard for public-key encryption and key establishment. It is efficient, secure, and is a leading candidate to replace the systems threatened by Shor's algorithm.
