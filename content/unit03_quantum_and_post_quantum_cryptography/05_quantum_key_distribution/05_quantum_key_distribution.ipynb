## 5.1 Core principle of QKD: secure key exchange using quantum mechanics.

Quantum Key Distribution (QKD) is a secure communication method that uses quantum mechanics to establish a shared, secret key between two parties (Alice and Bob) that can then be used to encrypt and decrypt messages.

The Core Principle: The security of QKD relies on a fundamental aspect of quantum mechanics: the act of measuring a quantum system can disturb it. If a third party, an eavesdropper (Eve), tries to intercept and measure the qubits being exchanged, her measurement will inevitably introduce detectable anomalies. This allows Alice and Bob to know if their key has been compromised. QKD is not for sending data itself, but for creating a key to secure data sent through other means.

## 5.2 Steps of the BB84 protocol.

The BB84 protocol, named after its creators Charles Bennett and Gilles Brassard (1984), is the first and most well-known QKD protocol.

### Protocol Steps
1. Step 1: Alice Sends Photons (Quantum Channel):
    * Alice generates two random strings: a string of bits (the future key) and a string of bases (rectilinear + or diagonal x).
    * For each bit, she prepares a photon (qubit) with a specific polarization depending on the bit and her chosen basis. She then sends this stream of photons to Bob over the quantum channel.
        * Rectilinear Basis (+): 0 → horizontal polarization, 1 → vertical polarization.
        * Diagonal Basis (x): 0 → 45° polarization, 1 → 135° polarization.
2. Step 2: Bob Measures Photons (Quantum Channel):
    * For each photon he receives, Bob randomly chooses a basis (+ or x) to measure it in.
    * He records his basis choice and the resulting bit (0 or 1) for each measurement.
3. Step 3: Basis Reconciliation (Classical Channel):
    * Alice and Bob connect over a public classical channel.
    * For each photon, they compare the basis they used. They do not reveal the bit values.
    * They discard all measurements where their bases did not match. The remaining bits, where they used the same basis, become their sifted key. On average, this happens for 50% of the qubits.
4. Step 4: Error Estimation (Classical Channel):
    * Alice and Bob publicly compare a random subset of their sifted key bits.
    * They calculate the Quantum Bit Error Rate (QBER). If there is no eavesdropping or environmental noise, the QBER should be 0%.
    * If the QBER is above a predetermined threshold, they assume an eavesdropper is present and abort the entire process.
5. Step 5: Key Distillation (Post-Processing):
    * If the QBER is acceptably low, they discard the bits used for testing.
    * They then perform two classical procedures on the remaining key:
        * Error Correction: Removes discrepancies in their keys caused by noise.
        * Privacy Amplification: Reduces any partial information Eve might have gained, creating a final, shorter, but perfectly secret key.

## 5.5 How an eavesdropper (Eve) is detected in BB84.

The security of BB84 comes from the fact that Eve cannot measure the qubits without leaving a trace.

### The Intercept-Resend Attack
1. Eve Intercepts: Eve catches a photon sent by Alice.
2. Eve Measures: To read the information, she must measure it. However, she does not know which basis Alice used. She has to guess (+ or x).
3. The Disturbance:
    * 50% of the time, Eve guesses the correct basis. She measures the correct bit value and can send an identical photon to Bob. No error is introduced.
    * 50% of the time, Eve guesses the wrong basis. Her measurement changes the photon's state. For example, if Alice sent a horizontal (0 in + basis) and Eve measures in the x basis, the photon is now either 45° or 135°.
4. Eve Resends: Eve sends the new, potentially altered, photon to Bob.
5. Errors are Created: Now, even if Bob uses the correct original basis (+), his measurement of the altered photon has a 50% chance of being incorrect (e.g., measuring vertical instead of horizontal).

**Detection:** This process introduces errors into the sifted key. When Alice and Bob compare a subset of their keys, they will find these mismatches. Eve's attempt to listen in raises the QBER, signaling her presence and causing Alice and Bob to discard the key.
